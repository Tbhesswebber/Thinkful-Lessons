```python
df2 = df[['5k','10k', '20k','half','25k', '30k', '35k','40k', 'overall', 'division','genderdiv','age', 'gender',  'official',
        'country', 'pace', 'state',  
          'city']]

df2.head()
```

```python
times = ['5k','10k', '20k','half','25k', '30k', '35k','40k']

for time in times:
    df2 = df2[df2[time] != '-']
    df2[time] = df2[time].astype(float)
```

```python
countries = df2['country'].unique()
for country in countries:
    singledout = df2[df2['country']== country]
    plt.scatter(df2['5k'], df2['overall'], c = 'blue', label = 'All')
    plt.scatter(singledout['5k'], singledout['overall'], c = 'yellow', label = country)
    plt.legend()
    plt.show()
    print('There are a total of {} entries from {}'.format(singledout.shape[0], country))
```

```python
male = df2[df2['gender']== 'M']
female = df2[df2['gender']== 'F']

for time in times:
    x = time
    y = 'age'
    plt.title('{} vs Age'.format(time))
    plt.scatter(male[x], male[y], c = 'Blue', label = 'Mens')
    plt.scatter(female[x], female[y], c = 'red', label = 'Womens')
    plt.legend()
    plt.show()
```

```python
df2.head()
test = df2[['5k','overall']]
```

```python
from sklearn.cluster import SpectralClustering

sc = SpectralClustering()
sc.fit(test)

#Predicted clusters.
predict=sc.fit_predict(test)

#Graph results.
plt.scatter(test[:, 0], test[:, 1], c=predict)
plt.show()
```
